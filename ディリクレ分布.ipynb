# ディリクレ分布
+ ベータ分布の多次元版
+ 制約: $\sum_{d=1}^{D}x_{d}=1$
+ $x_{d} \in R$
+ $\boldsymbol{x} = \set{x_{1}, x_{2}, \cdots, x_{D}}$
+ $\alpha_{d} \in R^{+}$
+ $\boldsymbol{\alpha} = \set{\alpha_{1}, \alpha_{2}, \cdots, \alpha_{D}}$

```math
\begin{matrix}
Dir(\boldsymbol{x}|\boldsymbol{\alpha}) = \frac{1}{D(\alpha)}\Pi_{d=1}^{D}x_{d}^{\alpha_{d}-1}
\end{matrix}
```

````python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
dir_dist = stats.dirichlet([.5, .5, .5])

# サンプリング
N = 1000
X = dir_dist.rvs(N)

# 散布図
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], alpha=.1)
plt.show()
````

様々なディリクレ分布

````python
alpha_list = [
    [0.1,0.1,0.1],
    [1.0,1.0,1.0],
    [5.0,5.0,5.0],
    [0.1,1.0,5.0],
]

fig, axes = plt.subplots(2, 2, figsize=(12,12))
for (i, ax) in enumerate(axes.ravel()):
    alpha = alpha_list[i]
    dir_dist = stats.dirichlet(alpha)
    X = dir_dist.rvs(N)
    ax.scatter(X[:,0], X[:,1], alpha=0.1)
    ax.set_title(f"alpha={alpha}")
plt.show()
````
