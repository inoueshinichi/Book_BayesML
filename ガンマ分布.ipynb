# ガンマ分布
+ $x \in R^{+}$
+ 実数の分布
+ 形状パラメータ: $\alpha \in R^{+}$
+ スケールパラメータ: $\theta \in R^{+}$
+ $\alpha=1$ の時, 指数分布

```math
\begin{matrix}
Gamma(x | \alpha, \theta) = \frac{1}{\Gamma(\alpha)\theta^{\alpha}}x^{\alpha-1}\exp{-\frac{x}{\theta}}
\end{matrix}
```


````python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
alpha = 1.0 # 形状パラメータ for ガンマ関数
theta = 1.0 # スケールパラメータ

gamma_dist = stats.gamma(a=alpha, scale=theta)

# サンプリング
N = 1000
X = gamma_dist.rvs(N)

# ヒストグラム
plt.hist(X, bins=10)
plt.show()
````

````python
# 確率密度関数
ls = np.linspace(0, 3, 100)
plt.plot(ls, gamma_dist.pdf(ls)) # α=1なので, これは指数分布
plt.show()
````

````python
ls = np.linspace(0,5,100)

alpha_list = [0.5,1.0,2.0]
theta_list = [0.5,1.0,2.0]

for alpha in alpha_list:
    for theta in theta_list:
        gamma_dist = stats.gamma(a=alpha, scale=theta)
        plt.plot(ls, gamma_dist.pdf(ls), label=f"alpha={alpha}, theta={theta}")

plt.legend()
plt.show()
````
