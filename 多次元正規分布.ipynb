# 多次元正規分布
```math
N(\boldsymbol{x} | \boldsymbol{\mu}, \boldsymbol{\Sigma}) = \frac{1}{\sqrt{(2\pi)^{D}|\Sigma|}}\exp(-(\boldsymbol{x}-\mu)^{T}\Sigma^{-1}(\boldsymbol{x} - \boldsymbol{\mu}))
```

````python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
mu = [0, 0]
cov = [
    [1, 0],
    [0, 1]
]

mvn_dist = stats.multivariate_normal(mu, cov)

# サンプリング(1000個)
N = 1000
X = mvn_dist.rvs(N)

# 散布図
plt.scatter(X[:,0], X[:,1])
plt.show()
````

````python
mu = [0, 0]
cov = [
    [1, .5],
    [.5, 1]
]

mvn_dist = stats.multivariate_normal(mu, cov)

# サンプリング(1000個)
N = 1000
X = mvn_dist.rvs(N)

# 散布図
plt.scatter(X[:,0], X[:,1])
plt.show()
````

等高線

````python
x1, x2 = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x1,x2))
plt.contourf(x1, x2, mvn_dist.pdf(pos))
````
