# 回帰モデル
+ 関数を生成する統計モデルのこと
+ 回帰モデルを使った教師あり学習を行うことは, 関数に対する事前分布を設計することに一致する

## 回帰モデルの例
1. データを生成する代わりに関数のパラメータを生成する方法
2. ノンパラメトリックベイズモデルであるガウス過程などを使うことで関数の特性を直接記述する方法

+ 関数のパラメータを生成する例
  
```math
\begin{matrix}
y &=& ax + b \\
a &\sim& N(0, 1.0) \\
b &\sim& N(0, 1.0)
\end{matrix}
```

````python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
````

ガウス分布による回帰モデル

````python
N = 10

ls = np.linspace(-3, 3, 100)
plt.figure()
for i in range(N):
    a = stats.norm(0, 1).rvs()
    b = stats.norm(0, 1).rvs()
    plt.plot(ls, a*ls+b)
plt.show()
````

ガンマ分布による回帰モデル

```math
a \sim Gamma(1.0, 1.0)
```

````python
plt.figure()
for i in range(N):
    a = stats.gamma(1,1).rvs()
    b = stats.norm(0,1).rvs()
    plt.plot(ls,a*ls+b)
plt.show()
````

指数分布の予測に回帰モデルを使う
+ $\alpha \sim N(0, 1.0)$
+ $\beta \sim N(0, 1.0)$


````python
plt.figure()
for i in range(N):
    a = stats.norm(0, 1).rvs()
    b = stats.norm(0, 1).rvs()
    plt.plot(ls, np.exp(a*ls+b))
plt.show()
````
