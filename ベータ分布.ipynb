# ベータ分布

```math
\begin{matrix}
x \in R^{D} \\
x \in (0,1) \\
\alpha \in R^{+} \\
\beta \in R^{+} \\
Beta(x | \alpha, \beta) = \frac{1}{B(\alpha, \beta)}x^{\alpha-1}(1-x)^{\beta-1} \\
\end{matrix}
```


## ベータ関数
```math
\begin{matrix}
x \in R^{D} \\
x \in (0,1) \\
\alpha \in R^{+} \\
\beta \in R^{+} \\
B(\alpha, \beta) = \int_{0}^{1}x^{\alpha-1}(1-x)^{\beta-1}dx
\end{matrix}
```

## ベータ関数とガンマ関数の関係
```math
\begin{matrix}
B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}
\end{matrix}
```

````python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
a=0.1
b=0.1
beta_dist = stats.beta(a,b)

# サンプリング
N = 1000
X = beta_dist.rvs(N)

# ヒストグラム
plt.hist(X, bins=10)
plt.show()
````

````python
# 確率密度関数
ls = np.linspace(0, 1, 100)
plt.plot(ls, beta_dist.pdf(ls))
````

````python
# 様々なベータ分布の形
# αとβの大きさで両端[0,1]を上から抑えるイメージ

ls = np.linspace(0,1, 100)

alpha_list = [0.1, 1.0, 2.0]
beta_list = [0.1, 1.0, 2.0]
for alpha in alpha_list:
    for beta in beta_list:
        beta_dist = stats.beta(alpha, beta)
        plt.plot(ls, beta_dist.pdf(ls), label=f"alpha = {alpha}, beta = {beta}")
        plt.legend()
        plt.tight_layout()
plt.show()
````
